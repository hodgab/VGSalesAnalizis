### Modulok importálása

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

### Adattábla beolvasása

```python
df = pd.read_csv("vgsales.csv")
```

```python
df.dropna(inplace = True)
```

```python
df['Year'] = df['Year'].astype(int)
```

```python
platform_sales = df.groupby('Platform')['Global_Sales'].sum().reset_index() # platform eladások csoportosítása a globális eladások alapján
top_platforms = platform_sales.sort_values(by='Global_Sales', ascending=False).head(10) # top 10 platform kiválasztása
```

### Top 10 platform a játék eladásokat tekintve

```python
# oszlop diagram design-ja és megjelenítése
sns.set_style('whitegrid')
sns.set_palette('deep')
sns.barplot(x='Platform', y='Global_Sales', data=top_platforms)
plt.title('Top 10 platform a játék eladásokat tekintve')
plt.xlabel('Platform')
plt.ylabel('Eladások (Millió dollárban)')
plt.show()
```

### A top 10 kiadó az eladásokat tekintve

```python
publishers = df.groupby('Publisher')['Global_Sales'].sum().reset_index() # kiadók csoportosítása a globális eladások alapján
```

```python
top_publishers = publishers.sort_values(by='Global_Sales', ascending = False).head(10) # top 10 kiadó kiválasztása
```

```python
# oszlopdiagram design-ja és megjelenítése
sns.set_style('darkgrid')
sns.set_palette('dark')
sns.barplot(x='Global_Sales', y='Publisher', data=top_publishers)
plt.title('Top 10 kiadó az eladások alapján')
plt.xlabel('Bevétel (Millió dollárban)')
plt.ylabel('Kiadó neve')
plt.show()
```

### A Nintendo és az EA eladásai a top 10-hez viszonyítva (százalékban)

```python
total_sales = top_publishers['Global_Sales'].sum() # a top 10 kiadó globális eladásai együttvéve
```

```python
publisher_perct = round((top_publishers['Global_Sales'] / total_sales * 100), 2) # a 10 kiadó eladásai a teljes globális eladáshoz nézve (%)
```

```python
# kördiagramm design-ja és megjeleítése
colors = plt.cm.Pastel1(range(len(publisher_perct)))
colors = [(r, g, b, 0.9) for (r, g, b, _) in colors]
explode = (0.05, 0.05, 0, 0, 0, 0, 0, 0, 0, 0)
textprops = {'color': 'black'}
plt.pie(publisher_perct, labels = ['Nin', 'EA', 'Act', 'SCE', 'Ubi', 'TTI', 'THQ', 'KDE', 'Sega', 'NBG'] , autopct='%1.1f%%', explode=explode, shadow=True, textprops=textprops, colors = colors)

plt.show()
```

```python
genre_sales = df.groupby(['Genre'])['Global_Sales'].sum() # globális bevételek csoportosítása kategóriák szerint
```

```python
genre_sales = genre_sales.sort_values(ascending=False) # a kategóriák csökkenő sorrendbe állítása
genre_sales = genre_sales.reset_index()
```

### Rangsorolás a kategóriákon belüli eladások szerint

```python
# oszlopdiagramm design-ja és megjelenítése
sns.set_style('whitegrid')
sns.set_palette('deep')
sns.barplot(x='Genre', y='Global_Sales', data=genre_sales)
plt.title('Rangsorolás a kategóriákon belüli eladások szerint')
plt.xlabel('Műfaj')
plt.xticks(rotation=90)
plt.ylabel('Eladások (Millió dollárban)')
plt.show()
```

```python
# a régiónkénti eladások százalékban a teljes globális eladáshoz viszonyítva
region_sales = np.empty((0,))
regions = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']
labels = ['North America', 'Europe', 'Japan', 'Other']
for region in regions:
    region_sale = (df[region].sum() / df['Global_Sales'].sum())
    region_sale = round(region_sale, 2)
    region_sales = np.append(region_sales, region_sale)
```

### Videójáték eladások régiónként

```python
# kördiagramm design-ja és megjelenítése
explode = (0.1, 0, 0, 0)
colors = sns.color_palette('pastel')[0:len(region_sales)]
plt.pie(region_sales, labels=labels, colors=colors, startangle=90, pctdistance=0.7, autopct='%1.1f%%', labeldistance=1.05,wedgeprops=dict(width=0.6, edgecolor='w'), shadow=True, explode=explode)
plt.show()
```

```python
# videójátékok globális eladásának a változása az évek alatt
sales_by_year = df.groupby(['Year'])['Global_Sales'].sum()
relevant_sales_by_year = sales_by_year[:-3]
```

```python
# vonaldiagramm megjelenítése és design-ja
sns.lineplot(data=relevant_sales_by_year)
sns.set_context("talk")
sns.despine(left=True, bottom=True)

max_index = relevant_sales_by_year.idxmax()
max_value = relevant_sales_by_year[max_index]

plt.annotate(f'Letöbb eladás: {max_value}',
             xy=(max_index, max_value),
             xytext=(max_index-1, max_value+2),
             arrowprops=dict(facecolor='black',arrowstyle='->'),
             bbox=dict(boxstyle='round', pad=0.2, fc='yellow', alpha=0.5))

plt.title("Videójáték eladások az évek alatt")
plt.xlabel("Év (1980-2015)")
xticks = np.arange(1980, 2016, 5)
plt.xticks(xticks)
plt.ylabel("Eladások (millió dollár)")
plt.show()
```

```python
# a Nintendo és az EA globális eladásainak változása az évek alatt
# Nintendo
nintendo = df[df['Publisher'] == 'Nintendo']
nintendo_sales_ot = nintendo.groupby(nintendo['Year'])['Global_Sales'].sum()
nintendo_sales_ot = nintendo_sales_ot[:-2]
# EA
ea = df[df['Publisher'] == "Electronic Arts"]
ea_sales_ot = ea.groupby(ea['Year'])['Global_Sales'].sum()
ea_sales_ot = ea_sales_ot[:-2]
```

```python
# a vonaldiagramm megjelenítése és design-ja
plt.plot(nintendo_sales_ot.index, nintendo_sales_ot, color="green", linewidth=2, label='Nintendo')
plt.plot(ea_sales_ot.index, ea_sales_ot, color="blue", linewidth=2, label='EA')
plt.legend()
plt.title("A Nintendo és az EA eladásai az évek alatt")
plt.xlabel("Év (1983-2015)", fontsize=14, fontweight='bold')
plt.ylabel("Bevétel (millió dollárban)", fontsize=14, fontweight='bold')
plt.rcParams['axes.facecolor'] = '#f0f0f0'
plt.grid(color='white', linestyle='--', linewidth=0.5)
xticks = np.arange(1983, 2016, 6)
plt.xticks(xticks)
plt.show()
```
